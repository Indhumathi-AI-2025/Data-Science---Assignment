##STEP 3: Handling the Missing Values

```python
#Finding the NaN Values in the Dataset

raw_dataset.isnull().sum()
```

```python
raw_dataset.fillna(0, inplace = True)
```

```python
raw_dataset.isnull().sum()
```

```python
#Remove the Redundant Columns in the Dataset (i.e, Zomato URL is not necessary)
raw_dataset.drop(['Zomato URL'], axis = 1, inplace = True)
raw_dataset
```

##STEP 4: Data Cleaning

```python
#Changed the Data Type of Dining Rating and Delivery Rating to Float Type
raw_dataset['Dining Rating'] = raw_dataset['Dining Rating'].astype(float)
raw_dataset['Delivery Rating'] = raw_dataset['Delivery Rating'].astype(float)
raw_dataset 
```

```python
raw_dataset.info()
```

```python
##To remove Square Brackets and Comma in Cuisine, Top Dishes, and Features

raw_dataset['Cuisine'] = raw_dataset['Cuisine'].replace("["," ")
raw_dataset['Cuisine'] = raw_dataset['Cuisine'].replace("]"," ")
raw_dataset['Cuisine'] = raw_dataset['Cuisine'].replace("'","")
raw_dataset['Top Dishes'] = raw_dataset['Top Dishes'].replace("["," ")
raw_dataset['Top Dishes'] = raw_dataset['Top Dishes'].replace("]"," ")
raw_dataset['Top Dishes'] = raw_dataset['Top Dishes'].replace("'","")
raw_dataset['Features'] = raw_dataset['Features'].replace("["," ")
raw_dataset['Features'] = raw_dataset['Features'].replace("]"," ")
raw_dataset['Features'] = raw_dataset['Features'].replace("'","")
raw_dataset
```

```python
#To remove duplicates

duplicate = raw_dataset.duplicated().sum()
print(duplicate)
```

```python
# Dropping all the duplicate rows
raw_dataset = raw_dataset.drop_duplicates()
raw_dataset
```
